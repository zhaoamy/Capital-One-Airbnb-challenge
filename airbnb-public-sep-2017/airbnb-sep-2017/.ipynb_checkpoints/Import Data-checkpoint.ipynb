```python
for neighbourhood in neighbourhood_boundaries.index:
    neighbourhood_boundaries.set_value(neighbourhood, 'min lat', (neighbourhood_coordinates.loc[neighbourhood])['latitude'].min())
    neighbourhood_boundaries.set_value(neighbourhood, 'max lat', (neighbourhood_coordinates.loc[neighbourhood])['latitude'].max())
    neighbourhood_boundaries.set_value(neighbourhood, 'min long', (neighbourhood_coordinates.loc[neighbourhood])['longitude'].min())
    neighbourhood_boundaries.set_value(neighbourhood, 'max long', (neighbourhood_coordinates.loc[neighbourhood])['longitude'].max())
```
